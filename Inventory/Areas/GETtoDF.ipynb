```python
import pip._vendor.requests
import json
import pandas as pd

URL = "https://api.assetfuture.com/inventory/Areas"   #Change URL if you want to query a different table
#URL = "https://api.assetfuture.com/model/Models"
Subscription_Key = '67ca2f9d707e44cd91a31ff3a06a15b'  
HEADERS = {'Cache-Control': 'no-cache','Ocp-Apim-Subscription-Key': Subscription_Key}

RECORDCOUNT = 3000    # The number of rows will be retrieved each time
STARTINDEX = 0


modelCount = RECORDCOUNT

df = pd.DataFrame()  # Create dataframe for final result

while modelCount >= RECORDCOUNT:
        print('Retrieving records: ',str(STARTINDEX),"-",str((STARTINDEX + RECORDCOUNT-1)))
        PARAMS = {'count': RECORDCOUNT,'startIndex': STARTINDEX}
        responseModels = pip._vendor.requests.get(url = URL, params = PARAMS, headers = HEADERS)
        models = responseModels.json()
        STARTINDEX = STARTINDEX + RECORDCOUNT
        modelCount = models["count"]
        data = json.loads(json.dumps(models))
        df2 = pd.json_normalize(data['data'])
        df = df.append(df2)    # Append the result of a single query to the final result dataframe

print('Transmission Completed. ' + str(STARTINDEX - RECORDCOUNT + modelCount) + ' records received')
```
